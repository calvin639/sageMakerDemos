<h3>Column Inspection</h3> <a class="anchor" id="c2.1"></a> </br>
It is necessary to drop a lot of the columns becasue useless information will slow down training and could corupt the model or distort the results. 
<img src="img/rubishModel.png" alt="Rubish Model" width="520" height="200"/>

```python
allCols=fifa.columns.tolist()
print("There are a total of {x} columns in the data\n".format(x=len(allCols)))
goodCols=allCols.copy()
dropCols=[]
#Lets save the players names for later
names=fifa['short_name']
#Should delete anything that relates to FIFA URLs
for c in goodCols:
    if "_url" in c:
        dropCols.append(c)
        
#Drop colums with name or ID
for c in ["sofifa_id","short_name","long_name"]:
    dropCols.append(c)

#It could be argued by certain teams playing styles that this could impact position but for now I'm going to remove team/nationality
for c in goodCols:
    if("club" in c) or ("nation" in c) or ("league" in c):
        dropCols.append(c)

#Now remove them         
for c in list(set(dropCols)):
    goodCols.remove(c)
print("There are a total of {x} columns left in the data\n".format(x=len(goodCols)))
```

#### Review
So there are still a lot of columns left and some of them are in my opinion going either unhelpful or maybe too directly related to the persons position, for example "cf" is their exact skill as a center forward.<br/>

```python
dropCols=['preferred_foot', 'overall','dob','potential','weak_foot','release_clause_eur','body_type','wage_eur','work_rate','player_tags','player_traits','real_face','ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram', 'lm', 'lcm', 'cm', 'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb', 'gk']
for c in dropCols:
    goodCols.remove(c)
print("There are a total of {x} columns left in the data\n".format(x=len(goodCols)))
```

<h3> Nulls</h3> <a class="anchor" id="c2.3"></a>

```python
print("Now lets deal with null values")
nulls=data.isnull().sum()/len(data)
nulls=pd.DataFrame(nulls, columns=["percentage"],index=nulls.index)
nulls=nulls.iloc[list(nulls['percentage'] > 0.0)]
nulls
```

From that simple analysis I think the best corse of action is:
1. Drop values with null value_eur because it is less than .4%. Normally not acceptable but its such a small amount.
2. Null goalkeeping_speed is probably every non-gaolkeeper in the dataset, which is a lot. Since they are not goalkeepers I will give them a value worse than the lowest speed in the dataset 
3. The remaining values it would make sence to give them the median value 

```python
#Drop null values for value_eur
data=data.dropna(subset=['value_eur'])
#find the worst gaolkeeping_speed
gkSpeeds=data['goalkeeping_speed'].iloc[list(data['goalkeeping_speed'].notnull())]
gkSpeeds=gkSpeeds.describe()
print(gkSpeeds)
#Give non-keepers a speed that is 3 standard deviations less than the mean. Also lower than slowest actual keepers value
slowKeeper=gkSpeeds['mean']-3*gkSpeeds['std']
data=data.fillna(value={'goalkeeping_speed':slowKeeper})
#Finally fill the remaining columns that contain nulls with the median values in that column 
medianCols=list(nulls.index)
medianCols.remove('goalkeeping_speed')
medianCols.remove('value_eur')
data=data.fillna(value=data[medianCols].median())
```

```python
nulls=data.isnull().sum()/len(data)
nulls=pd.DataFrame(nulls, columns=["percentage"],index=nulls.index)
nulls=nulls.iloc[list(nulls['percentage'] > 0.0)]
nulls
```

<h3>Training</h3> <a class="anchor" id="c3.2"></a>

#### Now lets run the KNN algorithm!
<aside style="float:right;margin-right:150px;">
    <img src="img/knn.png" alt="KNN" width="220" height="180" />
</aside>
K-Nearest Neighbors is a machine learning technique and algorithm that can be used for both regression and classification tasks. K-Nearest Neighbors examines the labels of a chosen number of data points surrounding a target data point, in order to make a prediction about the class that the data point falls into. <br>
I also considered using the K-Means algo becasue I think it could do a good job of classifying the data since we know the number of groups but for the sake of doing hyperparameter tuning with a test-accuracy I needed a supervised technique. 


```python
import matplotlib.pyplot as plt

import sagemaker
from sagemaker import get_execution_role
from sagemaker.serializers import CSVSerializer
from sagemaker.deserializers import JSONDeserializer
from sagemaker.amazon.amazon_estimator import get_image_uri

######
# Name : trained_estimator_from_hyperparams
######
# Def  : Train a KNN model 
######
# Args : - s3_train_data - Path to S3 training data
# Args : - hyperparams - Params for the KNN model
# Args : - output_path - S3 path where model will be saved 
# Args : - s3_test_data - Optional Path to S3 training data
######
# Name : trained_estimator_from_hyperparams
def train_knn_from_hyperparams(s3_train_data, hyperparams, output_path, s3_test_data=None):
    """
    Create an Estimator from the given hyperparams, fit to training data,
    and return a deployed predictor

    """
    # set up the estimator
    knn = sagemaker.estimator.Estimator(
        get_image_uri(boto3.Session().region_name, "knn"),
        get_execution_role(),
        instance_count=1,
        instance_type="ml.m5.2xlarge",
        output_path=output_path,
        sagemaker_session=sagemaker.Session(),
    )
    knn.set_hyperparameters(**hyperparams)

    # train a model. fit_input contains the locations of the train and test data
    fit_input = {"train": s3_train_data}
    if s3_test_data is not None:
        fit_input["test"] = s3_test_data
    knn.fit(fit_input)
    return knn
```

```python
#We can revisit the hyperparameters later by using Sagemakers hyper parameter tuning jobs
hyperparams = {"feature_dim": 45, "k": 10, "sample_size": 2000, "predictor_type": "classifier"}
output_path = f"s3://{bucket}/{prefix}/default_example/output"
knn_estimator = train_knn_from_hyperparams(
    s3_train_data, hyperparams, output_path, s3_test_data=s3_test_data
)
```

<h3>Testing</h3> <a class="anchor" id="c3.1"></a>
<br>
Lets deploy a prediction endpoint and use the test data to evaluate our model.

```python
import time

######
# Name : fifa_knn_predictor
######
# Def  : Make a prediction of a players position using KNN 
######
# Args : - knn_estimator - Sagemaker trained knn model
# Args : - estimator_name - Name we give to our estimator 
# Args : - instance_type - Hardware Sagemaker will use for predictions
# Args : - endpoint_name - Sagemaker endoint name for logging
######
# Name : trained_estimator_from_hyperparams
def fifa_knn_predictor(knn_estimator, estimator_name, instance_type, endpoint_name=None):
    knn_predictor = knn_estimator.deploy(
        initial_instance_count=1, instance_type=instance_type, endpoint_name=endpoint_name
    )
    knn_predictor.serializer = CSVSerializer()
    knn_predictor.deserializer = JSONDeserializer()
    return knn_predictor

instance_type = "ml.c4.large"
model_name = "knn_%s" % instance_type
endpoint_name = "knn-ml-c4-large-%s" % (str(time.time()).replace(".", "-"))
print("setting up the endpoint..")
predictor = fifa_knn_predictor(
    knn_estimator, model_name, instance_type, endpoint_name=endpoint_name
)
```

##### Now lets test it out!
We now have an active endpoint hosted on dedicated hardware to make inference. <br>
To test out this model I'm going to use the starting 11 from my favourite team, Liverpool!

```python
batches = np.array_split(X_test, 100)
print(f"data split into 100 batches, of size {batches[0].shape[0]}.")

# obtain an np array with the predictions for the entire test set
start_time = time.time()
predictions = []
for batch in batches:
    result = predictor.predict(batch, initial_args={"ContentType": "text/csv"})
    cur_predictions = np.array(
        [result["predictions"][i]["predicted_label"] for i in range(len(result["predictions"]))]
    )
    predictions.append(cur_predictions)
predictions = np.concatenate(predictions)
run_time = time.time() - start_time

test_size = y_test.shape[0]
num_correct = sum(predictions == y_test)
accuracy = num_correct / float(test_size)
print("time required for predicting %d data point: %.2f seconds" % (test_size, run_time))
print("accuracy of model: %.1f%%" % (accuracy * 100))
```

<h2 style="text-align:center;"> <u>Hyperparameter tuning </u></h2> <a class="anchor" id="c4"></a>

<h3>Set Up</h3> <a class="anchor" id="c4.1"></a>

```python
from sagemaker.tuner import (
    IntegerParameter,
    CategoricalParameter,
    ContinuousParameter,
    HyperparameterTuner,
)

# set up the estimator
output_path='s3://sagemaker-us-east-1-022761975017/knn-cwright-example-2022-04-06/auto-tune/output'
knn = sagemaker.estimator.Estimator(
    get_image_uri(boto3.Session().region_name, "knn"),
    get_execution_role(),
    instance_count=1,
    instance_type="ml.m5.2xlarge",
    output_path=output_path,
    sagemaker_session=sagemaker.Session(),
)

#Choose hyperparams
#Fixed
hyperparams = {"feature_dim": 45, "predictor_type": "classifier"}
knn.set_hyperparameters(**hyperparams)
#Tune-able
hyperparameter_ranges = {
    "k": IntegerParameter(5, 20),
    "sample_size": IntegerParameter(2000,20000, scaling_type='Logarithmic') #make sue to use logarithmic scale for large vals
}

#Objective we will measure on
objective_metric_name = "test:accuracy"
objective_metric_func = "Maximize"
#Max jobs to run
max_jobs=1 #Running the tuning with just a single job first to make sure we don't get any errors
#Max parallel jobs to run - note tuning uses previous results so don't do too much in parallel.
max_parallel_jobs=2
tuner = HyperparameterTuner(
    knn,
    objective_metric_name,
    hyperparameter_ranges,
    objective_type=objective_metric_func,
    max_jobs=max_jobs,
    max_parallel_jobs=max_parallel_jobs,
)
```

<h3>PCA</h3> <a class="anchor" id="c4.4"></a>

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
pca3 = PCA(n_components=3)
principalComponents = pca3.fit_transform(X_test)

principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2', 'principal component 3'])
principalDf['position']=predictions
principalDf['label']=y_test
print("Total variance explained by PCA is: {tot}\nBreakdown:{each}".format(tot=sum(pca3.explained_variance_ratio_),each=pca3.explained_variance_ratio_))
```

```python
fig = plt.figure(figsize=(9,9))
axes = Axes3D(fig)
axes.set_title('PCA Representation', size=14)
axes.set_xlabel('PC1')
axes.set_ylabel('PC2')
axes.set_zlabel('PC3')
posColourDict={0.0:'Orange',1.0:'Green',2.0:'Purple',3.0:'Black'}
for pos in posColourDict:
    df=principalDf.loc[principalDf['position']==pos]
    axes.scatter(df['principal component 1'],df['principal component 2'],df['principal component 3'],c=posColourDict[pos], cmap = 'prism', s=10)
axes.legend(['Goal Keeper','Defender','Midfielder','Forward'])
```
